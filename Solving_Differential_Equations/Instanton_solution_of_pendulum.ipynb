## A brief introduction of the backgroud

### Soliton
To get familiar with instanton, we start with soliton. Soliton is a classical wave package that typically appears in fluid. It was first reported in 1834.

> I was observing the motion of a boat which was rapidly drawn along a narrow channel by a pair of horses, when the boat suddenly stopped â€“ not so the mass of water in the channel which it had put in motion; it accumulated round the prow of the vessel in a state of violent agitation,  then suddenly leaving it behind, rolled forward with great velocity, assuming the form of a large solitary elevation, a rounded, smooth and well-defined heap of water, which continued its course along the channel apparently without change of form or diminution of speed. I followed it on horseback, and overtook it still rolling on at a rate of some eight or nine miles an hour, preserving its original figure some thirty feet long and a foot to a foot and a half in height. Its height gradually diminished, and after a chase of one or two miles I lost it in the windings of the channel. Such, in the month of August 1834, was my first chance interview with that singular and beautiful phenomenon which I have called the Wave of Translation.[6]

A soliton or solitary wave is a particle-like wave packet that maintains its shape while it propagates. According to Ref.[5], a solition has 3 properties:

1. They are of permanent form
2. They are localized within a region
3. They can interact with other solitons, and emerge from the collision unchanged, except for a phase shift.

We are already aware that this solution has character of particles. And in quantum world, a wave function of a soliton solution is called **instanton**.


### Instanton
**An instanton is a quantum counterpart of soliton**. And it is easy to map the feature of soliton from classical wave to quantum wave. While the soliton has the feature of spatial localization, the instanton is everywhere constant except a small time segment, like a bounce, or excitation of the vacuum.

Instantons were introduced all most fifty years ago in the context of Yang-Mills theory [2] and in quantum world the soliton is finally being treated as a particle, or pesudo-particle, because of the three characters described above.

This idea very quickly found many applications even in quantum mechanics, where they describe the tunneling processes from one vacuum to another. This is also the case in Yang-Mills theories where they describe tunnelling processes between different degenerate vacuum which are topologically distinct[3]. It is also used to construct Phenomenological model for Hubbard model in condensed matter physics[4].

Consider a particle moving in this one dimentional double well potential. Where there are two minima $+a,-a$ in this potential $V(q)$.

### Feynman path integral

We use Feynman path integral formalism to study the transition posibility of the particle between two wells.

The Feynman integral says that the transition possibility of a particle from state $q_i$ to $q_j$ is contibuted by all the path started from $q_i$ to $q_j$ weighted by their action.
$$
âŸ¨q_i|e^{i \hat{H}t/\hbar}|q_jâŸ© = \int_{q(0) = q_i,q(t) = q_j}Dq \exp [i\frac{\int L(q,\dot{q})dt}{\hbar}]
$$

where $L(q,\dot{q})$ is the Lagrangian.

### Wick rotation

Typically, to find the solution using Feynman integral method, we use belowing procedures.


1.  Find the stationary path (saddle point solution) using Eular-lagrangian equation, that's because the action reach it's minimum at this path. And the path with larger action will provide an oscillating phase and , after integration over all the remote path, give weaker contribution.  
2.  Consider the contribution of path near saddle point solution to add the quantum fluctuation correction (since the first term corresponds to classical motion)

However, for this double well potential, it is hard to find it's classical path, since it is hard to imagine in classic world how a particle which is innitially at rest moves from one well to another.

We perform wick rotation to solve this problem.
$$
 \tau â†’it
$$
And we have
$$
  L = m\dot{q}^2/2 - V(q) â†’ L = - m\dot{q}^2/2 - V(q)
$$

The equation of motion becomes
$$
-m\ddot{q} + V'(q) = 0
$$
as a consequence of the Wick rotation, there is an effective inversion of the potential, $V â†’âˆ’V$, The potential becomes

The classical solution of the Euclidean equation of motion that connects the two potential maxima is called an instanton solution while a solution traversing the same path but in the opposite direction ($â€œâˆ’a â†’ aâ€ â†’ â€œa â†’âˆ’aâ€$) is called an anti-instanton.

This solution has a time structure, it is almost everywhere constant except a small region of variation. These states is a well localized excitations and be treated as particles according to standard theorical field practice.

The classic action of a single instanton can be calculated:
$$
S = \int_0^\tau d\tau' (\frac{m\dot{q}^2}{2} + V(q)) =  \int_0^\tau d\tau' (m\dot{q}^2) = \int_0^\tau d\tau' (\frac{dq}{d\tau'}m\dot{q}) = \int_{-a}^adq \sqrt{2mV(q)}
$$
In the equation we use the condition
$$
\frac{m\dot{q}^2}{2} = V(q)
$$

The action of a instanton is solely determined by the profile of function $V(q)$

## instanton in pendulum

the potential felt by a pendulum is $-\cos \theta, \theta \in [-\pi/2,\pi/2]$. **This potential is similar to attractive region of the inverse of double well potential aboveï¼Œ Thus we expect instanton solution to this problem**

We need to solve the differential equation:
$$
\ddot{\theta} = k\sin \theta
$$
where $k$ refer to the accceleration of gravity. The angle between the vertical and the pendulum has been shifted by Ï€. We use Rungeâ€“Kutta to solve this equation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def driven_pendulum(x_initial,v_initial,tau_here,time_steps,k):
    t_pen=[0]
    v_pen=[v_initial]
    x_pen=[x_initial]
    tau=tau_here
    for i in range(1,time_steps):
        t_pen.append(tau*i)
        k1=tau*v_pen[i-1]
        l1=tau*(k*np.sin(x_pen[i-1]))
        k2=tau*(v_pen[i-1]+0.5*l1)
        l2=tau*(k*np.sin(x_pen[i-1]+0.5*k1))
        k3=tau*(v_pen[i-1]+0.5*l2)
        l3=tau*(k*np.sin(x_pen[i-1]+0.5*k2))
        k4=tau*(v_pen[i-1]+l3)
        l4=tau*(k*np.sin(x_pen[i-1]+k3))
        xnew=x_pen[i-1]+(1/6)*(k1+2*k2+2*k3+k4)
        # if xnew>np.pi:
        #     xnew=xnew-2*np.pi
        # if xnew<-np.pi:
        #     xnew=xnew+2*np.pi
        
        x_pen.append(xnew)
        v_pen.append(v_pen[i-1]+(1/6)*(l1+2*l2+2*l3+l4))
    return t_pen,v_pen,x_pen
```

```python
t_list,v_list,x_list=driven_pendulum(x_initial=1e-1,v_initial=0,tau_here=0.1,time_steps=100000,k = 0.0003)
```

```python
plt.figure(figsize=(20,7))
plt.plot(t_list,np.cos(x_list))
plt.xlabel('t',size=16)
plt.ylabel('h',size=16)
plt.title('$k=1,  x(0)=1e-1$',size=16)
```

As we tune the initial value of x, while we reaching $x = 0$, the feature of soliton becomes clear. This means that the particle is quite near from the peak of potential and the initial acceleration is very small. So the particle will spend more time in the peak.
